```python
import random

target = random.randint(1, 100)
# print(target)

print('--------------------------')
print('NUMBER GUESSING GAME\n')

#  Enter username and difficulty
username = input('Please enter your name:')
difficulty = input('Choose a difficulty: (Easy, Medium, or Hard)').lower().strip()
print(f'Game difficulty: {difficulty.upper()}')
print(f'Welcome', username)

def save_score(username, attempts, difficulty):
    with open('high_scores.txt', 'a') as file:
        file.write(f'{username}: {attempts} ({difficulty})\n')

def guess_checker(target, attempts, wrong_guesses):
    while attempts > 0:
        guess = int(input('Enter a number between 0 and 100'))

        if guess < target:
            print(f'{guess} is wrong! The target is higher')
        elif guess > target:
            print(f'{guess} is wrong! The target is lower')
        elif guess == target:
            print(f'{guess} is Correct! Congratulations!')
            save_score(username, attempts, difficulty) 
            break
        else:
            print('enter a valid number')
        
        attempts -= 1
        wrong_guesses += 1

        # Give a hint after 3 wrong guesses
        if wrong_guesses == 3:
            if target % 2 == 0:
                print("HINT: The number is EVEN")
            else:
                print("HINT: The number is ODD")

        print(f'You have {attempts} attempts remaining\n')
    else:
        print(f'Oops! You have run out of attempts. The number was {target}')
        save_score(username, attempts, difficulty)

# check difficulty and start game
if difficulty == 'easy':
    attempts = 10
    wrong_guesses = 0
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts, wrong_guesses)

elif difficulty == 'medium':
    attempts = 7
    wrong_guesses = 0
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts, wrong_guesses)

elif difficulty == 'hard':
    attempts = 5
    wrong_guesses = 0
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts, wrong_guesses)

else: 
    print('Invalid difficuly! Please enter a valid difficulty (Easy, Medium, or Hard)')
    
print('--------------------------')
```
