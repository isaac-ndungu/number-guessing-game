```python
import random

target = random.randint(1, 100)
print(target)

print('--------------------------')
print('NUMBER GUESSING GAME\n')

difficulty = input('Choose a difficulty: (Easy, Medium, or Hard)').lower().strip()
print(f'Game difficulty: {difficulty.upper()}')


def guess_checker(target, attempts):
    while attempts > 0:
        guess = int(input('Enter a number between 0 and 100'))

        if guess < target:
            print(f'{guess} is wrong! The target is higher')
        elif guess > target:
            print(f'{guess} is wrong! The target is lower')
        elif guess == target:
            print(f'{guess} is Correct! Congratulations!')
            break
        else:
            print('enter a valid number')
        
        attempts -= 1
        print(f'You have {attempts} attempts remaining\n')
    else:
        print('Oops! You have run out of attempts')

# check difficulty and start game
if difficulty == 'easy':
    attempts = 10
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts)

elif difficulty == 'medium':
    attempts = 7
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts)

elif difficulty == 'hard':
    attempts = 5
    print(f'You have {attempts} guesses\n')
    guess_checker(target, attempts)

else: 
    print('Invalid difficuly! Please enter a valid difficulty (Easy, Medium, or Hard)')
       

print('--------------------------')
```
